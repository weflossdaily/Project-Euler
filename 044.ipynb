# 44
Pentagonal numbers are generated by the formula $P_{n}=\frac{n\left(3n-1\right)}{2}.$ The first ten pentagonal numbers are: $1, 5, 12, 22, 35, 51, 70, 92, 117, 145, \ldots.$ It can be seen that $P_{4} + P_{7} = 22+70 = 92 = P_{8}.$ However, their difference, $70-22=48,$ is not pentagonal.

Find the pair of pentagonal numbers, $P_{j}$ and $P_{k},$ for which their sum and difference are pentagonal and $D = \left|P_{k}-P_{j}\right|$ is minimized. What is the value of $D$?

# Thoughts

## Made up words and notation abuse
  * pentagonal pair: a pair of pentagonal numbers $(P_{j},P_{k}),$ such that $j\neq k$
  * pentagonal difference: the difference between a members of a pentagonal pair, $\left|P_{j}-P_{k}\right|$
  * pentagonal sum: like pentagonal difference, but different somewhat
  * window: an index and an "offset," $(j,n),$ that correspond to a pentagonal pair, i.e., $(P_{j},P_{k=j+n})$
  * offset: the difference between sequence-terms' indexes, i.e., $n=\left|j-k\right|$ from the pentagonal pair $(P_{j},P_{k}),$ taken from the sequence of pentagonal numbers $P=p_{1}, p_{2}, \ldots,$ where all the notations $p_{n} = P_{n} = P(n)$ mean the same thing

Find a sequence of pairs $(j_{i},n_{i})$ such that the sequence
$D=d_{1},d_{2},\ldots,$
where
$d_{i} = n_{i}\left(6j_{i}+3n_{i}-1\right),$
is non-decreasing.

The term $d_{i}$ is the "pentagonal difference" for the $i^{\mathrm{th}}$ window. So if we iterate over this sequence and find the corresponding $(j,n)$ pairs, we will be "told" what the next "pentagonal pair" to test should be.

## Expressing Pentagonal Difference and Sum in Terms of Window Boundaries

### Difference
$\begin{eqnarray*}
 P(n)&=&\frac{n\left(3n-1\right)}{2}\\
 \Rightarrow P(k=j+n) - P(j) &=& \frac{(j+n)\left(3(j+n)-1\right)}{2} - \frac{j\left(3j-1\right)}{2}\\
 &=& \frac{1}{2}\left((j+n)\left(3(j+n)-1\right) - j\left(3j-1\right)\right) \\
 &=& \frac{1}{2}\left(3j^{2}+6jn+3n^{2}-j-n - 3j^{2} + j\right) \\
 &=& \frac{1}{2}\left(6jn+3n^{2}-n\right) \\
\Rightarrow d &=& \frac{1}{2}n\left(6j+3n-1\right)
\end{eqnarray*}$

### Sum
We express this in terms of the window $(j,n)$ to see if we can save some effort when determining if it's pentagonal.
$\begin{eqnarray*}
P(n)&=&\frac{n\left(3n-1\right)}{2}\\
 \Rightarrow P(k=j+n) + P(j) &=& \frac{(j+n)\left(3(j+n)-1\right)}{2} + \frac{j\left(3j-1\right)}{2}\\
  &=& \frac{1}{2}\left((j+n)\left(3(j+n)-1\right) + j\left(3j-1\right)\right) \\
  &=& \frac{1}{2}\left(3j^{2}+6jn+3n^{2}+j+n + 3j^{2} - j\right) \\
  &=& \frac{1}{2}\left(6j^{2}+6jn+3n^{2}+n\right) \\
  &=& 3j(j+n) + \frac{1}{2}n\left(3n+1\right)
\end{eqnarray*}$

## IsPentagonal check
$\begin{eqnarray*}
P(n)&=&\frac{n\left(3n-1\right)}{2}\\
 \Rightarrow 0 &=& 3n^{2}-n-2P_{n}\\
\Rightarrow n &=& \frac{-b \pm \sqrt{b^{2} - 4ac}}{2a} \\
\Rightarrow n &=& \frac{1 \pm \sqrt{1 - 4\cdot3\cdot(-2)P_{n}}}{6} \\
\end{eqnarray*}$

Since $P_{n}$ is positive, we can rule out one of the roots to get

$\begin{eqnarray*}
n &=& \frac{1 + \sqrt{1 + 24P_{n}}}{6} \\
\Rightarrow 6n-1 &=& \sqrt{1+24P_{n}} \\
\Rightarrow (6n-1)^{2} - 1 &=& 24P_{n}
\end{eqnarray*}$
