```python
from random import randint

jail = 10
goToJail = 30
cc1 = 2
cc2 = 17
cc3 = 33
ch1 = 7
ch2 = 22
ch3 = 36
c1 = 11
e3 = 24
h2 = 39
r1 = 5
r2 = 15
r3 = 25
r4 = 35
u1 = 12
u2 = 28

turns = 0
freq = [0] * 40
nextState = (0,None,None)

while turns < 10000000:
    turns += 1
    
    currentSpace, isLastRollDubs, isTwoRollsAgoDubs = nextState
    nextSpace = None
    
    die1 = randint(1,4)
    die2 = randint(1,4)
    
    doubles = None
    if die1 == die2:
        doubles = True
    else:
        doubles = False
        
    # trip dubs?
    if doubles == True and isLastRollDubs == True and isTwoRollsAgoDubs == True:
        nextSpace = jail
        nextState = (nextSpace,None,None) # resets doubles counter
        freq[jail] += 1
        continue
    
    # advance to rolled space
    roll = die1 + die2
    nextSpace = (currentSpace + roll) % 40
    
    # check for community chest
    if nextSpace == cc1 or nextSpace == cc2 or nextSpace == cc3:
        card = randint(1,16)
        if card == 1:
            # Advance to GO
            nextSpace = 0
        elif card == 2:
            # Go to jail
            nextSpace = jail
            nextState = (nextSpace,None,None) # resets doubles counter
            freq[jail] += 1
            continue
    
    # check for chance
    if nextSpace == ch1 or nextSpace == ch2 or nextSpace == ch3:
        card = randint(1,16)
        if card == 1:
            # Advance to GO
            nextSpace = 0
        elif card == 2:
            # Go to jail
            nextSpace = jail
            nextState = (nextSpace,None,None) # resets doubles counter
            freq[jail] += 1
            continue
        elif card == 3:
            nextSpace = c1
        elif card == 4:
            nextSpace = e3
        elif card == 5:
            nextSpace = h2
        elif card == 6:
            nextSpace = r1
        elif card == 7 or card == 8:
            # next railroad
            if nextSpace < r1:
                nextSpace = r1
            elif nextSpace < r2:
                nextSpace = r2
            elif nextSpace < r3:
                nextSpace = r3
            elif nextSpace < r4:
                nextSpace = r4
            else:
                nextSpace = r1
        elif card == 9:
            if nextSpace < u1:
                nextSpace = u1
            elif nextSpace < u2:
                nextSpace = u2
            else:
                nextSpace = u1
        elif card == 10:
            # Back three spaces, but land on CC
            if nextSpace == ch3:
                card = randint(1,16)
                if card == 1:
                    # Advance to GO
                    nextSpace = 0
                elif card == 2:
                    # Go to jail
                    nextSpace = jail
                    # nextState = (nextSpace,None,None) # resets doubles counter
                    nextState = (nextSpace,doubles,isLastRollDubs) # does not reset doubles counter
                    freq[jail] += 1
                    continue
                else:
                    nextSpace -= 3
            else:
                nextSpace -= 3
    
    nextState = (nextSpace,doubles,isLastRollDubs)
    freq[nextSpace] += 1
```
